```python
#Import file
import pandas as pd

file = 'textcollection.xlsx'
data = pd.read_excel(file, header=None)
```

```python
#Taking one column 
data = data[1]
data
```

```python
#Lowercasing
data = data.str.lower()
```

```python
#Stop Word Removal
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()
stopwords_list = pd.read_csv("stopwordbahasa.csv", engine='python', header=None)
stopwords = r'\b(?:{})\b'.format('|'.join(stopwords_list[0].values.tolist()))

data = data.str.replace(stopwords, '')
data
```

```python
#Stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

data = data.apply(lambda x: stemmer.stem(str(x)))

data = data.replace('-', "")
data = data.replace(['blank', []])
data = data.replace(['\t', []])

```

```python
#Tokenize
token_data = data.apply(lambda x:str(x).split())

token_data
```

```python
#Combine tokens
import collections

def flatten(x):
    result = []
    for el in x:
        if isinstance(x, collections.Iterable) and not isinstance(el, str):
            result.extend(flatten(el))
        else:
            result.append(el)
    return result

tokens = flatten(token_data)

tokens.sort()
tokens
```

```python
#Binary Term Doc Incidence
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

flatdata = flatten(data)

bin_vect = CountVectorizer(binary=True)
dicts_bin = bin_vect.fit_transform(flatdata).todense()
bin_vocab = bin_vect.vocabulary_

res = []
for i in token_data:
    k =  bin_vocab.keys()
    d = {}
    for j in k:
        d[j] = True if j in i else False
    res.append(d)

binary_vec = pd.DataFrame(res)
binary_vec['index'] = data.values
binary_vec = binary_vec.set_index('index')

pd.set_option('display.max_columns', 600)

binary_vec
```

```python
# Count Feature Extraction
   
vectorizer = CountVectorizer()
vec = vectorizer.fit_transform(flatdata)
vec1 = vec.toarray()
vec_df = pd.DataFrame(vec1)
vec_df['index'] = data.values
vec_df = vec_df.set_index('index')
vec_vocab = vectorizer.vocabulary_
vec_vocab = vec_vocab.keys()
vec_vocabs = sorted(vec_vocab)


vec_df.columns = vec_vocabs

vec_df
```

```python
#TFIDF

from sklearn.feature_extraction.text import TfidfVectorizer

tvectorizer = TfidfVectorizer()

tfidf = tvectorizer.fit_transform(flatdata).toarray()


vect_df = pd.DataFrame(tfidf)
vect_df['index'] = data.values
vect_df = vect_df.set_index('index')
vect_vocab = tvectorizer.vocabulary_
vect_vocab = vect_vocab.keys()
vect_vocabs = sorted(vect_vocab)


vect_df.columns = vect_vocabs

vect_df
```

```python
tfidf = tvectorizer.idf_

diction = dict(zip(tvectorizer.get_feature_names(), tfidf))
diction
```
